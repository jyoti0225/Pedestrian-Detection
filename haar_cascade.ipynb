```python
#Import libraries
import cv2 
import imutils
import numpy as np
from imutils.object_detection import non_max_suppression
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
# Initializing the HOG person detector 
hog = cv2.HOGDescriptor() 
hog.setSVMDetector(cv2.HOGDescriptor_getDefaultPeopleDetector()) 
```

```python
def show_image(img):
    cv2.imshow('Image' , img)
    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

```python
# Reading the Image 
image = cv2.imread('people_walking.jpg') 

# Resizing the Image 
image = imutils.resize(image, width=min(400, image.shape[1])) 
orig = image.copy()   
show_image(orig)
```

```python
#Display Image
plt.figure(num = 1, figsize=(8,8))
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.title('Original Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```

```python
# Detecting all the regions in the  
# Image that has a pedestrians inside it 
# detect people in the image
person_cascade = cv2.CascadeClassifier('haarcascade_fullbody.xml')
frame = cv2.imread('image2.jpeg')
# frame = cv2.resize(frame,(840,860)) # Resize to improve frame rate
gray_frame = cv2.cvtColor(frame, cv2.COLOR_RGB2GRAY) # Haar-cascade classifier needs a grayscale image
rects = person_cascade.detectMultiScale(gray_frame)
# print('recr',rects)
# draw the bounding boxes
for (x, y, w, h) in rects:
    cv2.rectangle(frame, (x,y), (x+w,y+h),(0,255,0),2)
    cv2.putText(image,'People',(x, y),cv2.FONT_HERSHEY_SIMPLEX,1,(255,23,0),1,cv2.LINE_AA)
# cv2_imshow(frame)
show_image(frame)
```

```python
#Display output
plt.figure(figsize=(8,8))
plt.imshow(cv2.cvtColor(frame, cv2.COLOR_BGR2RGB))
plt.title('Output Image', fontsize=26)
plt.gca().axes.get_yaxis().set_visible(False)
plt.gca().axes.get_xaxis().set_visible(False)
plt.show()
```
